Давайте научимся оценивать количество детей ceb по остальным признакам.

Загрузите данные и внимательно изучите их. Сколько разных значений принимает признак religion?

Во многих признаках есть пропущенные значения. Сколько объектов из 4361 останется, если выбросить все, содержащие пропуски?

```python
data.dropna().shape
```

```python
data.head()
```

```python
data['nevermarr'] = [1 if np.isnan(i) else 0 for i in data['agefm']]
```

```python
data.head()
```

```python
data = data.drop(columns=['evermarr']);
```

```python
data['agefm'] = [0 if np.isnan(i) else i for i in data['agefm']]
```

```python
data['heduc'] = [-1 if (np.isnan(i) and j == 1) else i
                for i, j in zip(data['heduc'], data['nevermarr'])]
```

```python
data['heduc'].shape[0] - data['heduc'].dropna().shape[0]
```

```python
data['idlnchld_noans'] = [-1 if np.isnan(i) else i for i in data['idlnchld']]
data['heduc_noans'] = [-2 if np.isnan(i) else i for i in data['heduc']]
data['usemeth_noans'] = [-1 if np.isnan(i) else i for i in data['usemeth']]
```

```python
data = data.drop(columns=['idlnchld', 'heduc', 'usemeth']);
```

```python
data
```

```python
data.shape
```

```python
data = data.dropna(axis=0)
```

```python
4054*15
```

```python
ans = ''
for i in data.columns:
    ans+=' + '
    ans+=i
```

```python
ans
```

```python
import statsmodels.formula.api as smf
```

```python
m1 = smf.ols('ceb ~ age + educ + religion + knowmeth + agefm + urban + electric + radio + tv + bicycle + nevermarr + idlnchld_noans + heduc_noans + usemeth_noans', 
             data=data)
fitted = m1.fit()
print(fitted.summary())
```

p <= 0.05
